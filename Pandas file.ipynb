```python
import pandas as pd
from pathlib import Path

# File to Load (Correct the path as per your system environment)
school_data_load = Path(r"C:\Users\Alexander (A.J.)\Desktop\Resources\schools_complete.csv")
student_data_load = Path(r"C:\Users\Alexander (A.J.)\Desktop\Resources\students_complete.csv")

# Read in School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_load)
student_data = pd.read_csv(student_data_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name"])
```

```python
#District Summary
#Perform the necessary calculations and then create a high-level snapshot of the district's key metrics in a DataFrame.
#Include the following:

#Total number of unique schools

#Total students

#Total budget

#Average math score

#Average reading score

#% passing math (the percentage of students who passed math)

#% passing reading (the percentage of students who passed reading)

#% overall passing (the percentage of students who passed math AND reading)






# number of unique schools
total_schools = school_data['school_name'].nunique()

# Calculating total number of students
total_students = school_data_complete["Student ID"].nunique()  

# Calculating the total budget
total_budget = school_data['budget'].sum()  # Added this calculation

# Calculating average scores
average_math_score = school_data_complete['math_score'].mean()
average_reading_score = school_data_complete['reading_score'].mean()

# Calculating percentage of students who passed math (70 or greater)
passing_math_count = school_data_complete[school_data_complete['math_score'] >= 70].count()['student_name']
percentage_passing_math = (passing_math_count / float(total_students)) * 100

# Calculating percentage of students who passed reading (70 or greater)
passing_reading_count = school_data_complete[school_data_complete['reading_score'] >= 70].count()['student_name']
percentage_passing_reading = (passing_reading_count / float(total_students)) * 100

# Calculating the percentage of students who passed both math and reading
overall_passing_count = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)].count()['student_name']
percentage_overall_passing = (overall_passing_count / float(total_students)) * 100

# Create a new dataframe to save results
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "Percentage Passing Math": [percentage_passing_math],
    "Percentage Passing Reading": [percentage_passing_reading],
    "Percentage Overall Passing": [percentage_overall_passing]
})

# Format
district_summary['Total Students'] = district_summary['Total Students'].map("{:,}".format)
district_summary['Total Budget'] = district_summary['Total Budget'].map("${:,.2f}".format)
district_summary['Average Math Score'] = district_summary['Average Math Score'].map("{:.1f}".format)
district_summary['Average Reading Score'] = district_summary['Average Reading Score'].map("{:.1f}".format)
district_summary['Percentage Passing Math'] = district_summary['Percentage Passing Math'].map("{:.1f}%".format)
district_summary['Percentage Passing Reading'] = district_summary['Percentage Passing Reading'].map("{:.1f}%".format)
district_summary['Percentage Overall Passing'] = district_summary['Percentage Overall Passing'].map("{:.1f}%".format)

district_summary

```
